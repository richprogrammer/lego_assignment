```python
add_parts = pd.merge(add_inventory_parts,parts,on = 'part_num')
add_parts.head()
```

```python
all_data = pd.merge(add_parts,colors,left_on = 'color_id', right_on='id' )
#all_data.head()
all_data.rename(columns={'name_x':'name_part','name_y':'name_color'}, inplace=True)
all_data.head()
```

```python
def set_parts(set_num, version):
    set_version= all_data[(all_data['set_num'] == set_num) & (all_data['version'] == version)]
    part_col = ['set_num', 'version', 'name_sets', 'year', 'name_themes', 'part_num', 'name_part', 'name_color', 'quantity', 'is_spare']
    return set_version[part_col].sort_values(['part_num', 'name_color']).reset_index(drop=True)
    
```
